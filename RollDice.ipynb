## read from csv convert into numpy array that show the data in seaborn 

```python
df = pd.read_excel('Marks.xlsx')
data = df.melt(id_vars='Name',var_name='Subject', value_name='Marks')
plt.figure(facecolor='black')
plt.subplot(111,facecolor='lightblue')
sns.barplot(data=data,x='Name',y='Marks',hue='Subject')
plt.title("Friends Marks",color='red')
plt.tick_params(colors='red')
plt.ylabel("Marks",color='red')
plt.show()
```
